```python
# 

import numpy as np

t = np.arange(0, 5, 0.01)

plt.figure(figsize=(10,12))

plt.subplot(411)
plt.plot(t, np.sqrt(t))
plt.grid()

plt.subplot(423)
plt.plot(t, t**2)
plt.grid()

plt.subplot(424)
plt.plot(t, t**3)
plt.grid()

plt.subplot(413)
plt.plot(t, np.sin(t))
plt.grid()

plt.subplot(414)  # 4 of 4x1 
plt.plot(t, np.cos(t))
plt.grid()

plt.show()
```
